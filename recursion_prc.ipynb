```python
def reverse_str(s):
  if len(s) == 0:
    return ""
  return s[-1] + reverse_str(s[:-1])

print(reverse_str("hello"))
```

```python
def count_occ(lst,target,index):
  count = 0
  if index == len(lst):
    return 0
  elif lst[index] == target:
    count = 1
  return count + count_occ(lst,target,index+1)

print(count_occ([1,2,3,4,5,1,2,1],2,0))
```

```python
def find_max(lst,index =0):
  if index == len(lst)-1:
    return lst[index]

  elif lst[index] > find_max(lst,index+1):
    return lst[index]
  else:
    return find_max(lst,index+1)

print(find_max([1,2,3,4,5,6,7,8,9,120,11]))
```
