# **Task 02**

```python

def find_max(lst, index):
  if index == len(lst):
    return None

  max_value = find_max(lst, index + 1)

  if max_value is None or lst[index] > max_value:
    return lst[index]

  return max_value

lst = [1, 2, 3, 4, 5]
print(find_max(lst, 0))
```

```python

def count_occ(lst,target,index=0):
  if index == len(lst):
    return 0

  if lst[index] == target:
    return 1 + count_occ(lst,target,index + 1)
  else:
    return count_occ(lst,target,index+1)

lst = [1, 2, 3, 4,2,5,2,6,8,2,5,6,2,6,1,3]
target = 2
print(count_occ(lst,target))
```

# **Task 12**

```python
def find_max(lst, index=0):
  if index == len(lst) - 1:
    return lst[index]
  sub_max = find_max(lst, index + 1)
  return lst[index] if lst[index] > sub_max else sub_max
nums = [8, 3, 15, 2, 9]
print(find_max(nums))

# find_max(4) → 9
# find_max(3) → max(2, 9)  = 9
# find_max(2) → max(15, 9) = 15
# find_max(1) → max(3, 15) = 15
# find_max(0) → max(8, 15) = 15      # 15 must be answer
```

# **Task 15**

```python
def count_occ(lst, target, index=0):
  if index == len(lst):
    return 0
  return (1 if lst[index] == target else 0) + count_occ(lst, target, index+1)
print(count_occ([1,2,1,3,1,4], 1))

# index 5: lst[5] = 4 → not equal to 1 → 0 + 0 = 0
# index 4: lst[4] = 1 → equal to 1 → 1 + 0 = 1
# index 3: lst[3] = 3 → not equal to 1 → 0 + 1 = 1
# index 2: lst[2] = 1 → equal to 1 → 1 + 1 = 2
# index 1: lst[1] = 2 → not equal to 1 → 0 + 2 = 2
# index 0: lst[0] = 1 → equal to 1 → 1 + 2 = 3

# Final result: 3
```

# **Task 16**

```python
count = 0
def fib(n):
  global count
  count += 1
  if n <= 1:
    return n
  return fib(n-1) + fib(n-2)
print(fib(4))
print("Calls:", count) # 9 calls and 3 is output
```
