# **Task 01**

```python
def find_item(data, target):
    if data == target:
        return []

    if type(data) == dict:
        for key, value in data.items():
            result = find_item(value, target)
            if result is not None:
                return [key] + result

    elif type(data) == list:
        for item in data:
            result = find_item(item, target)
            if result is not None:
                return result
    return None

data = {
    "Fruits": {"Citrus": ["Orange", "Lemon"], "Berries": ["Strawberry"]},
    "Vegetables": {"Roots": ["Carrot"], "Leafy": ["Spinach"]}
}

print(find_item(data, "Lemon"))
```

# **Task 02**

```python
def count_nodes(tree):

    if not tree[1]:
        return 1

    total = 1

    for child in tree[1]:
        total += count_nodes(child)

    return total

tree = ("root", [
    ("docs", [("file1", []), ("file2", [])]),
    ("images", []),
    ("code", [("main.py", [])])
])
print(count_nodes(tree))
```

# **Task 03**

```python
def flatten(items):
    flat_list = []

    for item in items:
        if type(item) == list:
            flat_list.extend(flatten(item))
        else:
            flat_list.append(item)

    return flat_list
items = [1, [2, 3, [4]], 5, [6, [7, 8]]]
print(flatten(items))
```

# **Task 04**

```python
def freq_recursive(nums, index=0, freq={}):
    if index >= len(nums):
        return freq

    num = nums[index]
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

    return freq_recursive(nums, index + 1, freq)
nums = [1, 2, 2, 3, 1, 4, 4, 4]
print(freq_recursive(nums))
```

# **Task 05**

```python
def recursive_intersection(list_of_sets):
    if not list_of_sets:
        return set()
    if len(list_of_sets) == 1:
        return list_of_sets[0]
    return list_of_sets[0] & recursive_intersection(list_of_sets[1:])
sets_list = [{1,2,3}, {2,3,4}, {3,4,5}]
print(recursive_intersection(sets_list))
```

```python
def list_to_tuple(lst):
    if not lst:
        return ()
    return (lst[0],) + list_to_tuple(lst[1:])

print(list_to_tuple([1,2,3,4]))
```

# **Task 07**

```python
def remove_consecutive_duplicates(lst):
    if len(lst) <= 1:
        return lst
    if lst[0] == lst[1]:
        return remove_consecutive_duplicates(lst[1:])
    else:
        return [lst[0]] + remove_consecutive_duplicates(lst[1:])
print(remove_consecutive_duplicates([1,1,2,2,2,3,1,1]))
```

# **Task 08**

```python
def depth(nested_list):
    if type(nested_list) != list or not nested_list:
        return 0

    current_max = 0

    for item in nested_list:
        d = depth(item)

        if d > current_max:
            current_max = d

    return 1 + current_max

print(depth([1,[2,[3,[4]]]]))
```

# **Task 09**

```python
def combine_dicts(dicts, index=0, combined={}):
    if index >= len(dicts):
        return combined

    for key, value in dicts[index].items():
        if key in combined:
            combined[key] += value
        else:
            combined[key] = value

    return combine_dicts(dicts, index + 1, combined)
dicts = [
    {"A": 1},
    {"B": 2},
    {"A": 3},
    {"C": 1}
]
print(combine_dicts(dicts))
```

# **Task 10**

```python
def power_set(s):
    s = list(s)
    if not s:
        return [set()]
    rest = power_set(s[1:])
    result = []
    for r in rest:
        result.append(r)
        result.append(r | {s[0]})
    return result

print(power_set({1,2,3}))
```

# **TAsk 14**

```python
def count(d, keys):
  if not keys:
    return 0
  return d.get(keys[0], 0) + count(d, keys[1:])
d = {"x": 2, "y": 3, "z": 5}
print(count(d, ["x", "z", "a", "y"]))
```

# **Task 15**

```python
def mystery(d, keys):
  if keys == []:
    return []
  k = keys[0]
  value = d.get(k, 0)
  return [value * len(k)] + mystery(d, keys[1:])
data = {"a": 2, "bb": 3, "ccc": 1}
print(mystery(data, ["a", "ccc", "bb"]))
```
