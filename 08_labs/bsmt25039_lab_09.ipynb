```python
regions = ["North", "South", "East", "West"]
sales_data = [
    [1200, 1500, 1100],
    [1000, 950, 1050],
    [1600, 1700, 1650],
    [900, 800, 950]
]

print("Region | Total Sales | Average Sale")
print("---------------------------------------")

for i in range(4):
    total = 0
    for j in sales_data[i]:
        total += j

    average = total / len(sales_data[i])
    print(f"{regions[i]}  |      {total}  | {average:12.2f}")

    if average > 1500:
        print("Top performing region:", regions[i])

print("End of report")
```
