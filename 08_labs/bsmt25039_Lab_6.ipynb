# **Task 9**

```python
fruits = {"apple", "banana", "cherry", "mango", "orange"}
tropical = {"mango", "banana", "papaya"}
citrus = {"orange", "lemon", "lime"}

if tropical.isdisjoint(citrus):
  print("Tropical and citrus fruits have no overlap")
else:
  print("Some fruits are both tropical and citrus.")

if fruits.issuperset(tropical) and fruits.issuperset(citrus):
  print("The fruit list covers all tropical and citrus fruits.")
else:
  print("The fruit list is missing some fruits.")

if tropical.issubset(fruits) and not tropical.issubset(citrus):
  print("All tropical fruits are in the list but not all are citrus.")
else:
  print("Nothing")
```
