# ***Task 1***

```python
a= int(input("Enter your first number : "))

b = int(input("\nEnter your second number : "))

c = int(input("\nEnter your third number : "))

d = int(input("\nEnter your fourth number : "))

exp1_result = a + b * c - d

exp2_result = (a + b) * (c - d)

exp3_result = a + (b * (c- d))

exp4_result = (a + b * c) - d


print(f"\n{a} + {b} * {c} - {d} is : {exp1_result}")

print(f"\n({a } + {b} ) * ({c}- {d}) is : {exp2_result}")

print(f"\n { a} +  ({b } * ({c} - {d})) is : {exp3_result}")

print(f"\n({a} + {b} * {c}) - {d} is : {exp4_result}")
```

# **Task 3**

```python
value1 = int(input("Enter your first value : "))
value2 = int(input("Enter your second value : "))


and_use = value1 == value2 and value1 < value2

or_use = value1 > value2 or value1 < value2

not_use = not(value1 > value2)

print(f"\nThe result of and operation is : {and_use}")
print(f"\nThe result of or operation is : {or_use}")
print(f"\nThe result of not operation is : {not_use}")
```

# **Task 4**

```python
import math
num_1 = int(input("Enter your first number : "))
num_2 = int(input("Enter your second number : "))

check_is_greater= math.sqrt(num_1 ) > math.sqrt(num_2)

check_is_equal = math.sqrt(num_1) == math.sqrt(num_2)

check_not_equal = math.sqrt(num_1) != math.sqrt(num_2)

print(f"\nThe result of Sqrt {num_1} is greater than Sqrt {num_2} is  : {check_is_greater}")
print(f"\nThe result of Sqrt {num_1} is equal to Sqrt {num_2} is  : {check_is_equal}")
print(f"\nThe result of Sqrt {num_1} is not equal to Sqrt {num_2} is  : {check_not_equal}")
```

# **Task 5**

```python
factorial_input= int(input("Enter number you want to find factorial : "))

base_number_input = int(input("\nEnter your base number : "))

exponent_number_input = int(input("\nEnter your power number : "))

factorial = math.factorial(factorial_input)

power = math.pow(base_number_input, exponent_number_input)

print(f"\nThe factorial of {factorial_input} is : {factorial}")
print(f"\n{base_number_input} to the power of {exponent_number_input} is : {power}")

check_factorial_greater_power = factorial > power
check_both_are_equal = factorial == power

print(f"\nThe factorial of {factorial_input} is greater than the {base_number_input } to the power {exponent_number_input}: {check_factorial_greater_power}")
print(f"\nThe factorial of {factorial_input} is equal to {base_number_input} to the power {exponent_number_input} : {check_both_are_equal}")



```

# **Task 6**

```python

radius_input = float(input("Enter the radius : "))

circumference = 2 * math.pi * radius_input

area = math.pi * math.pow(radius_input, 2)

check_area_greater_50 = area > 50

print(f"\nThe radius of the circle is : {radius_input}")
print(f"\nThe circumference of the circle is : {circumference:.3f}")
print(f"\nThe area of the circle is : {area:.4f} and the area is greater than 50 : {check_area_greater_50}")
```
