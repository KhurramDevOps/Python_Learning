```python
my_list = [10,20,30,40,50,60]
max = my_list[0]
min = my_list[0]
i = 0
while  i < len(my_list):
  if my_list[i] > max:
    max = my_list[i]
  if my_list[i] < min:
    min = my_list[i]
  i+=1
  diff = max - min

print("The maximun number is ",max)
print("The minimun number is ",min)
print("The difference between max and min is",diff)
```

# **Task 3**

```python
tem_in_cel = []

while True:
  user = int(input("Enter temperature in celcius : "))
  if user == 999:
    break
  if user > -50 and user < 60:
     tem_in_cel.append(user)
  else:
    print("Temperature should be in range of -50 to 60")

maxi = max(tem_in_cel)
mini = min(tem_in_cel)
sum4 = sum(tem_in_cel)
avg = sum(tem_in_cel)/ len(tem_in_cel)

print("\nYou break the loop")
print("\nAll recorded tem are : ",tem_in_cel)
print("\nThe maximum temperature is ",maxi)
print("\nThe minimum temperature is ",mini)
print("\nThe sum of temperatures is ",sum4)
print("\nThe average temperature is ",avg)
```

# **Task 4**

```python
petrol_sold_in_litres = []
sales_over_50_litres = []
while True:
  liters = int(input("Enter liters of petrol sold : "))
  if liters == 0:
    break
  if liters > 50:
    sales_over_50_litres.append(liters)
  elif liters < 0:
    print("Invalid input")
  else :
    petrol_sold_in_litres.append(liters)

avy_liters = sum(petrol_sold_in_litres + sales_over_50_litres) / (len(petrol_sold_in_litres) + len(sales_over_50_litres))


print("\nThe petrol sold in liters are in each transaction",petrol_sold_in_litres)
print("\nThe petrol sold (above 50 liters) transactions are:", sales_over_50_litres)
print("\nNumber of pertol sold each transaction above 50 liters are : ",len(sales_over_50_litres))
print("\nThe average liters of petrol sold is ",avy_liters)

if sum(petrol_sold_in_litres + sales_over_50_litres) > 500 :
  print("High Sales Day")

else :
  print("Normal Sales Day")
```

# **Task 5**

```python
students = []
unique_students = set()

while True:
  user = input("Enter STUDENTS name who are in class : ").lower()

  if user == "end":
    break
  else :
    students.append(user)
    unique_students.add(user)

print("\nThe students in class are : ",students)
print("\nThe unique students in class are : ",unique_students)
print("\nThe total number of unique students who attend class are : ",len(unique_students))
print("\nTotal number of duplicates students are " , len(students) - len(unique_students))
```

# **Task 6**

```python
friend_1 = []
friend_2 = []

while True:
  user_1 = int(input("Enter favourite number of friend 1 : "))
  if user_1 == -1:
    break
  else:
    friend_1.append(user_1)

unique_friend_1 = set(friend_1)
print("\n")
while True:
  user_2 = int(input("Enter favourite number of friend 2 : "))
  if user_2 == -1:
    break
  else:
    friend_2.append(user_2)

unique_friend_2 = set(friend_2)

print("\nThe unique favourite numbers of friend 1 are : ",unique_friend_1)
print("\nThe unique favourite numbers of friend 2 are : ",unique_friend_2)
print("\nThe common favourite numbers are : ",unique_friend_1 & unique_friend_2)
print("\nThe friends who have no common favourite numbers are : ",unique_friend_1 ^ unique_friend_2)
print("\nThe total shared nummber are : ",len(unique_friend_1 & unique_friend_2))
```

```python

total_spent = 0

while True:
  input_amount = int(input("Enter the amount of the order : "))
  if input_amount == 0:
    break
  if input_amount < 0:
    print("Invalid input")
    continue
  total_spent += input_amount


if total_spent > 100000:
  shopper_category = "Premium Customer"
elif total_spent > 30000 and total_spent < 100000:
  shopper_category = "Regular Customer"
elif total_spent < 30000:
  shopper_category = "New Customer"


print(f"\nTotal amount of the order : {total_spent}")
print(f"Shopper category : {shopper_category}")

```

# **Task 8**

```python
CAR_TOLL = 200
BUS_TOLL = 500
TRUCK_TOLL = 800
BIKE_TOLL = 100

car_count = 0
bus_count = 0
truck_count = 0
bike_count = 0


vehicle_types = []

total_tolls = []
while True:
  vehicle_entered = input("ENter the vehicle type : ").lower().strip()

  if vehicle_entered == "stop":
    break

  if vehicle_entered == "car":
    total_tolls.append(CAR_TOLL)
    vehicle_types.append(vehicle_entered)
    car_count += 1

  elif vehicle_entered == "bus":
    total_tolls.append(BUS_TOLL)
    vehicle_types.append(vehicle_entered)
    bus_count += 1

  elif vehicle_entered == "truck":
    total_tolls.append(TRUCK_TOLL)
    vehicle_types.append(vehicle_entered)
    truck_count += 1

  elif vehicle_entered == "bike":
    total_tolls.append(BIKE_TOLL)
    vehicle_types.append(vehicle_entered)
    bike_count += 1

  else :
    print("Invalid input")


sum_toll = sum(total_tolls)

if total_tolls:
    max_toll = max(total_tolls)
    min_toll = min(total_tolls)
else:
    max_toll = min_toll = None



if sum_toll > 10000:
  traffic_day = "High Traffic Day"
elif sum_toll >= 5000 and sum_toll <= 10000:
  traffic_day = "Moderate Traffic Day"
else:
  traffic_day = "Low Traffic Day"


print("\nAll amount tolls collected list are : ",total_tolls)
print("\nThe total amount of tolls is : ",sum_toll)
print("\nThe Highest amount of toll is : ",max_toll)
print("\nThe Lowest amount of toll is : ",min_toll)
print("\nThe unique vehicles entered are : ",set(vehicle_types))
print("\nThe total number of cars paid toll are : ",car_count)
print("\nThe total number of buses paid toll are : ",bus_count)
print("\nThe total number of trucks paid toll are : ",truck_count)
print("\nThe total number of bikes paid toll are : ",bike_count)
print("\nToday is :",traffic_day)
```
