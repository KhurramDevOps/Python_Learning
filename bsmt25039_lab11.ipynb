# **Task 05**

```python
def next_run(start_index, nums):
    run = [nums[start_index]]
    for i in range(start_index + 1, len(nums)):
        if nums[i] > run[-1]:
            run.append(nums[i])
        else:
            break
    return run

def collect_increasing_runs(nums):
    all_runs = []
    current_index = 0
    while current_index < len(nums):
        run = next_run(current_index, nums)
        all_runs.append(run)
        current_index += len(run)
    return all_runs


print(collect_increasing_runs([3,5,7,2,4]))
```

# **Task 08**

```python
no_repeat = []

def unique_words(sentence):
  words = sentence.split()

  for i in words:
    if i in no_repeat:
      no_repeat.remove(i)
    else:
      no_repeat.append(i)
  repeat = set(set(words) - set(no_repeat))
  print(f"Words that appear ones are : {no_repeat}")
  print(f"Words that repeat are : {repeat}")

unique_words("hello world hello  is programming is easy or is not python have easy syntax and python and ")


```

# **TAsk 09**

```python

def border_sum(matrix):

    total = 0
    rows = len(matrix)
    cols = len(matrix[0])

    for i in range(rows):
        for j in range(cols):
            if i == 0 or i == rows - 1 or j == 0 or j == cols - 1:
                total += matrix[i][j]

    return total

border_sum([[1,2,3],[4,5,6],[7,8,9]])
```

# **Task 10**

```python
def sort_by_frequency(nums):
    d = {}
    for n in nums:
        if n in d:
            d[n] += 1
        else:
            d[n] = 1
    items = sorted(d.items())

    def get_freq(t):
        return t[1]

    items = sorted(items, key=get_freq)
    return items

print(sort_by_frequency([4,4,1,2,2,3]))
```
