```python
def sum_digits(n):
  if n < 10 :
    return n

  return n%10 +sum_digits(n//10)

sum_digits(358)
```

```python
def reverse_str(s):
  if len(s) == 0:
    return s

  return s[-1] + reverse_str(s[:-1])
reverse_str("hello")
```

```python
count = 0
def count_occ(lst,target,index=0):
  global count
  if index == len(lst):
    return count

  if lst[index] == target:
    count += 1
  return count_occ(lst,target,index+1)


print(count_occ([1,2,3,4,3,2,3,2,1,4,3],3))
```

```python
count = 0
def count_fib_calls(n):
  global count
  count += 1

  if n == 0:
    return 0
  if n ==1 :
    return 1


  count_fib_calls(n-1) + count_fib_calls(n-2)
  return count

print(count_fib_calls(6))

```

```python
dic = {"total" : 0 , "above_80" : 0, "bonus_applied" : set() , "max_quiz" : 0}

def performance_analyzer(lst,index=0):
  global dic
  if index == len(lst):
    return dic

  dic["total"] += lst[index]
  if lst[index] > 80:
    dic["above_80"] += 1

  if lst[index] < 50:
    lst[index] = lst[index] + 10
    dic["bonus_applied"].add(index)

  if lst[index] > dic["max_quiz"]:
    dic["max_quiz"] = lst[index]

  return performance_analyzer(lst,index+1)

print(performance_analyzer([80,44,85,39,70]))

```

```python
highest_total_product = ""
count_low_stock = 0
unique_stock = set()
dic={}

def track_stock(lst,index=0):
  global dic, unique_stock, count_low_stock, highest_total_product

  if index == len(lst):
    return dic, unique_stock, count_low_stock, highest_total_product

  p,q = lst[index]
  if p in dic:
    dic[p] += q
  else:
    dic[p] = q

  unique_stock.add(p)


  for k , v in dic.items():
    if v < 20:
      count_low_stock +=1


  maximum = max(dic.values())
  for k,v in dic.items():
    if v == maximum:
      highest_total_product = k

  return track_stock(lst,index+1)


print(track_stock([("mango",30),("apple",19),("banana",60), ("mango",18)]))
```

```python
orders = [
    {"amount": 5000, "category": "electronics", "refund": False},
    {"amount": 12000, "category": "electronics", "refund": True},
    {"amount": 700, "category": "books", "refund": False},
    {"amount": 1500, "category": "clothing", "refund": True},
    {"amount": 20000, "category": "electronics", "refund": False},
    {"amount": 800, "category": "books", "refund": True},
    {"amount": 300, "category": "clothing", "refund": False},
    {"amount": 11000, "category": "clothing", "refund": False},
    {"amount": 2500, "category": "books", "refund": False},
    {"amount": 500, "category": "electronics", "refund": True}]


large_purchase = 0
total_revenue = 0
dic = {}
tuple_=("",0)

def order_check(lst,index=0):
  global large_purchase, dic, tuple_, total_revenue

  if index == len(lst):
    return dic, large_purchase, tuple_, total_revenue

  if lst[index]["refund"] ==False:
    total_revenue += lst[index]["amount"]

  if lst[index]["amount"] > 10000:
    large_purchase += 1
  if lst[index]["refund"] == True:
    if lst[index]["category"] in dic:
      dic[lst[index]["category"]] += 1
    else:
      dic[lst[index]["category"]] = 1


  for i in dic:
    if dic[i] > tuple_[1]:
      tuple_ = (i,dic[i])

  return order_check(lst,index+1)

print(order_check(orders))
```

```python
temperatures = [
    22.5, 23.0, 23.5, 24.0,   # increasing sequence (4 in a row)
    20.0, 19.5, 19.5,         # duplicate 19.5
    18.0, 18.2,
    21.0, 21.5, 21.5,         # duplicate 21.5
    21.0,                     # 21.0 repeats
    20.0, 20.5, 21.0,
    22.0, 23.5, 23.0, 22.0
]


def temp(lst,index=0,di=None,inc=0,total=0,seen=set()):
  if di==None:
    di= {
  "average": 0,                   # float
  "largest_drop": 0,                # float (magnitude)
  "duplicates_count": 0,              # int (see interpretation above)
  "duplicates_set": set(), # set of duplicate values
  "increasing_sequence": False}

  if index == len(lst):
    return di

  total += lst[index]
  di["average"] = total/len(lst)

  drop = lst[index-1] - lst[index]
  if drop > di["largest_drop"]:
    di["largest_drop"] += drop

  if lst[index] in seen:
      di["duplicates_set"].add(lst[index])
      di["duplicates_count"] += 1
  else:
      seen.add(lst[index])

  if lst[index] > lst[index-1]:
    inc += 1
    if inc >= 4:
      di["increasing_sequence"] = True
  else:
    inc = 0


  return temp(lst,index+1,di,inc,total,seen)

temp(temperatures)
```

```python
numbers = [1, [2, 3, [4]], 5, [6, [7, 8]]]
def items(lst):
  new_lst = []
  for num in lst:
      if type(num) == list:
        new_lst.extend(items(num))
      else:
        new_lst.append(num)

  return new_lst
print(items(numbers))
```

```python
def count_nodes(tree):

    if not tree[1]:
        return 1

    total = 1

    for child in tree[1]:
        total += count_nodes(child)

    return total

tree = ("root", [
    ("docs", [("file1", []), ("file2", [])]),
    ("images", []),
    ("code", [("main.py", [])])
])
print(count_nodes(tree))
```
