```python
quiz_marks = [45, 82, 76, 39, 91, 58, 47, 88]


def quiz(lst,index=0,di = {"total":0, "bonus_applied" : set() , "above_80" : 0 , "max_quiz" : 0}):
  if index == len(lst):
    return di



  if lst[index] < 50:
    lst[index] +=10
    di["bonus_applied"].add(lst[index])
  if lst[index] > 80:
    di["above_80"] += 1

  di["total"] += lst[index]

  if lst[index] > di["max_quiz"]:
    di["max_quiz"] = lst[index]

  return quiz(lst,index+1,di)

quiz(quiz_marks)

```

```python
stock_entries = [
    ("Rice", 50),
    ("Wheat", 18),
    ("Sugar", 30),
    ("Rice", 25),
    ("Oil", 15),
    ("Wheat", 40),
    ("Sugar", 10),
    ("Oil", 60),
    ("Rice", 12)
]

def stock(lst,index=0,di={"Rice":0,"Wheat":0,"Sugar":0,"Oil":0}, unique_product=set() , highest_total_product = "" , low_stock_product = 0 ):
  if index == len(lst):
    return f"{di} , unique products are : {unique_product} , product with highest total count : {highest_total_product} , number of product with count < 20  :  {low_stock_product}"

  product , quantity = lst[index]
  di[product] += quantity
  unique_product.add(product)

  if di[product] > all(di):
    highest_total_product = product

  if di[product] < 20:
    low_stock_product += 1

  return stock(lst,index+1,di,unique_product,highest_total_product,low_stock_product)

stock(stock_entries)

```

```python
hourly_steps = [
    0, 0, 0, 0, 0,
    300, 450, 600, 700, 800,
    900, 400, 0,
    600, 650, 700, 750,
    200, 0,
    550, 580,
    0, 0, 300
]



def steps(lst,index = 0,di = {"total_steps": 0,"longest_active_streak": 0,"skipped_hours": 0,"sleep_detected": False} ,consec_zero = 0 , consec_work=0):
  if index ==len(lst):
    return di


  di["total_steps"] += lst[index]

  if lst[index] == 0:
    consec_zero += 1
    di["skipped_hours"] += 1
    if consec_zero >= 5:
      di["sleep_detected"] = True

  else:
     consec_zero = 0


  if lst[index] > 500:
    consec_work +=1
    if consec_work > di["longest_active_streak"]:
      di["longest_active_streak"] = consec_work
  else:
    consec_work = 0

  return steps(lst,index+1,di,consec_zero,consec_work)

steps(hourly_steps)
```

```python
orders = [
    {"amount": 5000, "category": "electronics", "refund": False},
    {"amount": 12000, "category": "electronics", "refund": True},
    {"amount": 700, "category": "books", "refund": False},
    {"amount": 1500, "category": "clothing", "refund": True},
    {"amount": 20000, "category": "electronics", "refund": False},
    {"amount": 800, "category": "books", "refund": True},
    {"amount": 300, "category": "clothing", "refund": False},
    {"amount": 11000, "category": "clothing", "refund": False},
    {"amount": 2500, "category": "books", "refund": False},
    {"amount": 500, "category": "electronics", "refund": True}
]


def order(lst,index=0 , total_revenue=0 , refund_high = ("",0) , large_purchase = 0 , refund_count = {'electronics': 0, 'clothing': 0, 'books': 0} ):
  if index == len(lst):
    return total_revenue , refund_high , large_purchase , refund_count

  if lst[index]["refund"] == False:

    total_revenue += lst[index]["amount"]

  if lst[index]["amount"] > 10000:
    large_purchase += 1

  if lst[index]["refund"] == True:
    refund_count[lst[index]["category"]] += 1

  for i in refund_count:
    if refund_count[i] > refund_high[1]:
      refund_high = (i,refund_count[i])

  return order(lst,index+1,total_revenue,refund_high,large_purchase,refund_count)

order(orders)
```

```python
temperatures = [
    22.5, 23.0, 23.5, 24.0,   # increasing sequence (4 in a row)
    20.0, 19.5, 19.5,         # duplicate 19.5
    18.0, 18.2,
    21.0, 21.5, 21.5,         # duplicate 21.5
    21.0,                     # 21.0 repeats
    20.0, 20.5, 21.0,
    22.0, 23.5, 23.0, 22.0
]


def temp(lst,index=0,di=None,inc=0,total=0,seen=set()):
  if di==None:
    di= {
  "average": 0,                   # float
  "largest_drop": 0,                # float (magnitude)
  "duplicates_count": 0,              # int (see interpretation above)
  "duplicates_set": set(), # set of duplicate values
  "increasing_sequence": False}

  if index == len(lst):
    return di

  total += lst[index]
  di["average"] = total/len(lst)

  drop = lst[index-1] - lst[index]
  if drop > di["largest_drop"]:
    di["largest_drop"] += drop

  if lst[index] in seen:
      di["duplicates_set"].add(lst[index])
      di["duplicates_count"] += 1
  else:
      seen.add(lst[index])

  if lst[index] > lst[index-1]:
    inc += 1
    if inc >= 4:
      di["increasing_sequence"] = True
  else:
    inc = 0


  return temp(lst,index+1,di,inc,total,seen)

temp(temperatures)

```

```python
def temp(lst, index=0, di=None, inc=0, total=0, seen=None):
    if di is None:
        di = {
            "average": 0,
            "largest_drop": 0,
            "duplicates_count": 0,
            "duplicates_set": set(),
            "increasing_sequence": False
        }
    if seen is None:
        seen = set()

    if index == len(lst):
        di["average"] = total / len(lst)
        return di

    # Average
    total += lst[index]

    # Largest drop
    if index > 0:
        drop = lst[index-1] - lst[index]
        if drop > di["largest_drop"]:
            di["largest_drop"] = drop

        # Increasing sequence
        if lst[index] > lst[index-1]:
            inc += 1
            if inc >= 2:  # index-1 + index gives 3 consecutive increasing
                di["increasing_sequence"] = True
        else:
            inc = 0

    # Duplicates
    if lst[index] in seen:
        di["duplicates_set"].add(lst[index])
        di["duplicates_count"] += 1
    else:
        seen.add(lst[index])

    return temp(lst, index+1, di, inc, total, seen)


print(temp(temperatures))
```
