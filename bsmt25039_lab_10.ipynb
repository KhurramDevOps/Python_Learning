# **Task 03**

```python
dict_3 = {"a":1,"b":2,"c":3,"d":4,"e":5}

dict_3.update({"d":6}) # update function is use to update a value in the dictionary and to combine two dictionary

print(f"The updated dict using update is : {dict_3}")

dict_3.pop("e") # pop function is use to remove the key and its value if none given its give error

print(f"\nThe updated dict using pop is  : {dict_3}")

dict_3.clear()  # clear function is use to clear all the dictionary and after that empty dictionary will be printed

print(f"\nThe updated dict using clear is  : {dict_3}")
```

# **Task 04**

```python

dict_4 = {"a":1,"b":2,"c":3,"d":4,"e":5,"b":1 , "c" :7 , "g":4}

invert = {}

for key, val in dict_4.items():
  if val in invert:
    invert[val].append(key)
  else:
    invert[val] = [key]

print(invert)
```
