# **Task 02**

```python

student_marks = [85, 92, 78, 88, 33, 20 ,17 ,90, 76, 95, 89, 84, 91,48]
max = student_marks[0]
min = student_marks[0]

pass_count = 0
fail_count = 0

total = 0
for i in student_marks:
  if i > max:
    max = i
  if i < min:
    min = i
  total +=i
  average = total / len(student_marks)

  if i >= 50:
    pass_count += 1
  else:
    fail_count += 1


print("Highest mark is :" , max)
print("\nLowest mark is :" , min)
print("\nAverage mark is :" , average)
print("\nNumber of students passed is :" , pass_count)
print("\nNumber of students failed is :" , fail_count)
```

# **Task 05**

```python
temps = (30, 32, 33, 31, 29, 35, 36)

coldest_day = temps[0]
hottest_day = temps[0]

print("Below are the temperatures > 32 a day : ")
for i in temps:

  if i > 32:
    print(i)
  if i < coldest_day:
    coldest_day = i
  if i > hottest_day:
    hottest_day = i

print("Coldest day temperature is :" , coldest_day)
print("Hottest day temperature is :" , hottest_day)

```

# **Task 12**

```python
word_input = input("Enter a word to check repeated letters : ")

count_letters = {}

repeated = set()

print("Repeated letters in a word are below : ")

for ch in word_input:
  if ch in count_letters:
    repeated.add(ch)


  else:
    count_letters[ch] = 1

# print(count_letters)

print(", ".join(repeated))
```

# **Task 15**

```python
transactions = [500, -200, 1000, -150, 300, -400, 250]

deposits = 0
withdrawls = 0

for i in transactions:
  if i>0:
    deposits += i
  else:
    withdrawls += -i

print(f"Total deposits : {deposits}")
print(f"Total withdrawls : {withdrawls}")
print(f"Net balance is : {deposits - withdrawls}")
```
